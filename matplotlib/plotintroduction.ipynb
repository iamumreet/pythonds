```python
import matplotlib.pyplot as plt
# plt.xkcd()

# programming = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
# prices = [29.99,39.99, 24.99, 34.99, 19.99 ]
# plt.plot(programming, prices)

# plt.title('Book Prices')
# plt.xlabel('Books')
# plt.ylabel('Prices ($)')

emp_age = [25,30,35,40,45,50,55,60]
salary_range = [30000, 40000,50000, 60000, 70000, 80000, 90000, 100000]
plt.plot(emp_age, salary_range, marker = 'o', linestyle = '-', color = 'b')
python_salary = [32000, 42000, 52000, 30000, 72000, 82000, 92000, 102000]
plt.plot(emp_age, python_salary, marker = 's', linestyle = '--', color = 'r')
plt.title('Employee Age vs Salary')
plt.xlabel('Employee Age')
plt.ylabel('Salary ($)')
plt.legend(['General Salary', 'Python Salary'])
plt.show()

# label = "general salary", label = "python salary"
# plt.legend()
```
