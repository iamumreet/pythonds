```python
import matplotlib.pyplot as plt
Years = [2019, 2020, 2021, 2022, 2023]
Sales= [100, 120, 150, 170, 200 ]
plt.plot(Years, Sales ,marker = 'o', linestyle = '--', color = 'b')
plt.title('Total Sales Per Year')
plt.xlabel('Years')
plt.ylabel('Sales ($)')
# plt.gca().set_facecolor('lightgray')
# plt.grid(True, linestyle = '--', alpha = 0.5)
# plt.savefig('plot.png')
plt.show()

```

```python
import matplotlib.pyplot as plt

courses = ["Python", "Java", "C++", "JavaScript"]
students = [40, 25, 20, 35]
plt.figure(figsize=(8,5))
plt.bar(courses, students, color='red')
plt.title("Number of Students in Each Course")
plt.xlabel("Courses")
plt.ylabel("Number of Students")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Expense": [200, 250, 300, 280]
})
data

plt.figure(figsize=(8,5))
plt.plot(data["Month"], data["Expense"], marker='o', color='blue', linestyle='-')
plt.title("Monthly Expenses")
plt.xlabel("Month")
plt.ylabel("Expense")
plt.savefig('plot.png')
plt.grid(True)
plt.show()
```
