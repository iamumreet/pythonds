```python
import pandas as pd
data   = pd.read_csv("../data/students.csv")
# data.dropna()
# data.dropna(inplace = True)
data.fillna(76,inplace = True)
data['total']=data['eng']+data['mat']+data['sic']+data['com']+data['hel']
data['percentage'] = data['total']/5
def grade(value):
    if value>=90:
        return 'A'
    elif value>=80:
        return 'B'
    elif value>=70:
        return 'C'
    elif value>=60:
        return 'D'
    else:
        return 'F'
data['grade'] = data['percentage'].apply(grade)
data 
# data.info()
# data.describe()
```
